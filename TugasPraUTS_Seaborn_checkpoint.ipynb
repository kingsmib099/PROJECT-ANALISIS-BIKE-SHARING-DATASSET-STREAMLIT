### Pertanyaan 1:

Berapakah Jumlah Total Order Di Masing-Masing Bulan Pada Tahun 2011 - (10122483 - Muhammad Japa Maulana)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Membaca dataset
df = pd.read_csv('day.csv')

# Memfilter data untuk tahun 2011S
df_2011 = df[df['yr'] == 0]

# Menghitung jumlah total order di masing-masing bulan
total_order_setiap_bulan = df_2011.groupby('mnth')['cnt'].sum()

# Menampilkan hasil
total_order_setiap_bulan
```

```python
# Menampilkan Grafik Dengan Diagram
plt.figure(figsize=(10, 6))
total_order_setiap_bulan.plot(kind='bar', color='skyblue')
plt.title('Total Order per Bulan di 2011')
plt.xlabel('Bulan')
plt.ylabel('Total Order')
plt.xticks(rotation=0)
plt.show()
```

```python
# Menghitung jumlah total order di masing-masing bulan
total_order_setiap_bulan = df_2011.groupby('mnth')['cnt'].sum().reset_index()
```

```python
plt.figure(figsize=(8, 8))
colors = sns.color_palette('pastel')[0:len(total_order_setiap_bulan)]
plt.pie(total_order_setiap_bulan['cnt'],
        labels=total_order_setiap_bulan['mnth'],
        autopct='%1.1f%%',
        startangle=180,
        colors=colors,
        wedgeprops=dict(width=0.5))

# Menambahkan lingkaran di tengah sebagai donut chart
centre_circle = plt.Circle((0, 0), 0.70, fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

plt.title('Proporsi Total Pemesanan per Bulan pada tahun 2011')
plt.show()
```

```python

data = pd.read_csv("day.csv")

# Mengonversi kolom "dteday" ke dalam tipe data datetime
data['dteday'] = pd.to_datetime(data['dteday'])

# Membuat filter untuk tahun 2012
filter_2012 = (data['dteday'] >= '2012-01-01') & (data['dteday'] <= '2012-12-31')

# Mengambil data untuk tahun 2012 dan kolom "registered"
registered_2012 = data.loc[filter_2012, ['dteday', 'registered']]

registered_2012
```

```python
# Memilih data untuk tahun 2012
data_2012 = data.loc[filter_2012, ['dteday', 'registered']]

# Menambahkan kolom bulan untuk mengelompokkan berdasarkan bulan
data_2012['month'] = data_2012['dteday'].dt.month

# Mengelompokkan berdasarkan kolom bulan dan menjumlahkan nilai "registered" untuk setiap grup
total_registrasi_perbulan = data_2012.groupby('month').agg({'registered': 'sum'})

total_registrasi_perbulan
```

```python
import seaborn as sns

sns.set(style="whitegrid")
plt.figure(figsize=(10, 6))


# Plot batang berdasarkan 'dteday' dan 'registered'
sns.barplot( x ='month', y ='registered', hue = 'month' ,data=total_registrasi_perbulan, legend=False)

plt.title('Jumlah registerasi tahun 2012')
plt.xlabel('Bulan')
plt.ylabel('Jumlah Registrasi')

plt.show()
```

```python
import matplotlib.pyplot as plt

# Mengelompokkan berdasarkan kolom bulan dan menjumlahkan nilai "registered" untuk setiap grup
total_registrasi_perbulan = data_2012.groupby('month').agg({'registered': 'sum'}).reset_index()


# Membuat pie chart
plt.figure(figsize=(8, 8))

# Palet warna bawaan dari Matplotlib
colors = plt.cm.tab20.colors[:len(total_registrasi_perbulan)]

plt.pie(total_registrasi_perbulan['registered'], labels=total_registrasi_perbulan['month'], autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Persentasi Total registrasi pengguna perbulan pada Tahun 2012')
plt.show()
```
